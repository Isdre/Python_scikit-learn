```python
import pandas as pd
data = pd.read_csv("DisneylandReviews.csv", encoding='latin-1')
data.dropna(inplace=True)
data.head(5)
```

```python
data["Sentiment"] = "Neutral"
data.loc[data["Rating"] < 3, "Sentiment"] = "Negative"
data.loc[data["Rating"] > 3, "Sentiment"] = "Positive"
print(data.loc[data["Sentiment"] == "Positive"]["Sentiment"].count())
print(data.loc[data["Sentiment"] == "Neutral"]["Sentiment"].count())
print(data.loc[data["Sentiment"] == "Negative"]["Sentiment"].count())
```

```python
reviews = []
class Review:
    def __init__(self,text,sentiment):
        self.text = text
        self.sentiment = sentiment
for text,senti in zip(data["Review_Text"],data["Sentiment"]):
    review = Review(text,senti)
    reviews.append(review)
#     print(reviews[-1].text[0],reviews[-1].sentiment)
```
