```python
import pandas as pd
data = pd.read_csv("DisneylandReviews.csv", encoding='latin-1')
data.dropna(inplace=True)
data.head(5)
```

```python
data["Sentiment"] = "Neutral"
data.loc[data["Rating"] < 3, "Sentiment"] = "Negative"
data.loc[data["Rating"] > 3, "Sentiment"] = "Positive"
print(data.loc[data["Sentiment"] == "Positive"]["Sentiment"].count())
print(data.loc[data["Sentiment"] == "Neutral"]["Sentiment"].count())
print(data.loc[data["Sentiment"] == "Negative"]["Sentiment"].count())
data.head(5)
```

```python
reviews = []
class Review:
    def __init__(self,text,sentiment):
        self.text = text
        self.sentiment = sentiment
for text,senti in zip(data["Review_Text"],data["Sentiment"]):
    review = Review(text,senti)
    reviews.append(review)
#     print(reviews[-1].text[0],reviews[-1].sentiment)
reviews[:5]
```

```python
a = 5000#ile wierszy
from sklearn.model_selection import train_test_split
train,test = train_test_split(reviews[:a],test_size=0.2,random_state=50)
print(len(train))
print(len(test))
```

```python
x_train = [x.text for x in train]
y_train = [y.sentiment for y in train]

x_test = [x.text for x in test]
y_test = [y.sentiment for y in test]

y_test[0]
```

```python
#https://scikit-learn.org/stable/tutorial/text_analytics/working_with_text_data.html
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

train_x = vectorizer.fit_transform(x_train)
test_x = vectorizer.transform(x_test)

print(test_x[0])
```

```python
from sklearn import svm
clf_svc = svm.SVC(kernel="linear")

clf_svc.fit(train_x,y_train)
# test_x[0]
# prediction = clf_svc.predict(test_x)
print(clf_svc.score(test_x,y_test))
```

```python
import joblib
joblib.dump(clf_svc,"DisneyReviewer_SVC.joblib")
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_dec = DecisionTreeClassifier()
clf_dec.fit(train_x,y_train)
print(clf_dec.score(test_x,y_test))
```

```python
joblib.dump(clf_svc,"DisneyReviewer_DTC.joblib")
```

```python
from sklearn.naive_bayes import GaussianNB
clf_gnb = GaussianNB()
clf_gnb.fit(train_x.toarray(),y_train)
# prediction = clf_gnb.predict(test_x.toarray())
print(clf_gnb.score(test_x.toarray(),y_test))
```

```python
joblib.dump(clf_gnb,"DisneyReviewer_gnb.joblib")
```

```python
from sklearn.linear_model import LogisticRegression
clf_lr = LogisticRegression(max_iter=5000)
clf_lr.fit(train_x,y_train)
print(clf_lr.score(test_x,y_test))
```

```python
joblib.dump(clf_lr,"DisneyReviewer_LogisticRegression.joblib")
```
