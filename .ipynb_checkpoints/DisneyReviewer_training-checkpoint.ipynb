```python
import pandas as pd
reviews = pd.read_csv("DisneylandReviews.csv", encoding='latin-1')
reviews.head(5)
reviews.dropna(inplace=True)
len(reviews)
```

```python
reviews["Sentiment"] = "Neutral"
reviews.loc[reviews["Rating"] < 3, "Sentiment"] = "Negative"
reviews.loc[reviews["Rating"] > 3, "Sentiment"] = "Positive"
```
